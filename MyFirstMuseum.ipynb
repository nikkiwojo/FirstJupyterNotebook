```python
import pandas as pd
```

```python
museum_df = pd.read_csv('museums.csv')
museum_df
```

```python
# Which city or state has the most museums per capita?

state = museum_df['State (Administrative Location)']
city = museum_df['City (Administrative Location)']

def most_museums():
    if city.max() > state.max():
        return city.max()
    else:
        return state.max()

most_museums()
```

```python
# Which city or state has the lease museums per capita?

def least_museums():
    if city.min() < state.min():
        return city.min()
    else:
        return state.min()

least_museums()
```
